# Resampling the Data:

```python
sns.distplot(raw_data["fetal_health"])
```

The above distplot portrays an imbalance in the target variable 'fetal_health' which is explains more about normal inspection denoted as "1" in fetal_health rather than suspected and pathological catogorical attributes of fetal_health. To rectify this, a resampling technique is employed to achieve a more equitable representation of all three fetal health classes. Resampling techniques helps to prevent biased model while training and evaluating, especially when minority classes are underrepresented. After using resampling techniques, the distribution among classes is visually inspected to ensure a proper balance. This proactive approach to handling class imbalance enhances the reliability and fairness of predictive modeling. Overall, the utilization of resampling techniques demonstrates a commitment to improving the robustness of the analysis.

```python
from imblearn.over_sampling import SMOTE
```

```python
#Creating variables for target and independent data 
pred_data = raw_data.iloc[:, :-1]
target = raw_data.iloc[:, -1]
# Applying smooth to normalize the outcast data to avoid imbalance
smote = SMOTE(random_state=42)
# the smooth is applied over pred and target and associate it with new variables namely x_resampled and y_resampled
x_resampled, y_resampled = smote.fit_resample(pred_data, target)
#Concatinating bothe x_resampled,  y_resampled and target variable with the attached structure of the columns
resamp_data = pd.concat([pd.DataFrame(x_resampled, columns=pred_data.columns), pd.Series(y_resampled, name='target')], axis=1)

print('Total number of duplicate records:',resamp_data.duplicated().sum())
print('Total number of Null value records:',resamp_data.isnull().sum().sum())
```

After resampling the data 
The data explains a balanced output with 100% of data for all three categories

```python
onez=0
twoz=0
threez=0 
for i in resamp_data["target"]:
    if i == 1:
        onez+=1
    elif i==2:
        twoz+=1
    else:
        threez+=1

        
print(f" After resampling using the SMOTE the data is normalized which was in the form of imbalance state where :\n Normal={onez}\n Suspected={twoz}\n Pathological={threez} ")
    
```

```python

print(f"Shape before resampleing {raw_data.shape}\n After resampling {resamp_data.shape}")       
    
```

The diagram below displays the balanced output

```python

sns.set(style="whitegrid")
sns.histplot(data=resamp_data, x='target', bins=10, color='skyblue', edgecolor='black')
plt.xlabel('fetal health')
plt.ylabel('Frequency')
plt.title('Distribution of Fetal health')
plt.show()
```

# Normalizing Data:

RobustScaler was chosen for scaling due to its resilience to outliers. By using the median and interquartile range (IQR), it provides reliable standardization even in the presence of extreme values, enhancing machine learning algorithm performance

```python
from sklearn.preprocessing import RobustScaler 
```

```python
#Separating both data set for scaling to reduce the effect of the outliers in the data
blc_data = resamp_data[pred_data.columns]
blc_target = resamp_data['target']
```

```python
# Associating a variable called rob_scaler 
rob_scaler = RobustScaler()
scaled = rob_scaler.fit_transform(blc_data)
scaled_data = pd.DataFrame(scaled, columns=blc_data.columns)
```

```python
scaled_data.describe()
```

The code utilizes a RobustScaler method to perform robust scaling on independent predictor variables blc_data, which is done by standardizing data by removing the median and scaling based on the interquartile range (IQR). The fit_transform() method used above is specially used to compute the scaling of parameters and thereby applied for transformation. Resulting scaled data is stored in scaled_data, maintaining original column names. describe() provides summary statistics for the scaled data.


## Splitting the Dataset into Train and Test Sets

This code segment splits the scaled predictor variables (scaled_data) and corresponding target variable (blc_target) into training and testing sets using the train_test_split function from scikit-learn Library. The test_size parameter specifies the proportion of the dataset to include in the testing set, which is set to 30%. The random_state parameter ensures reproducibility of the split. Splitting the data into training and testing sets allows for model training on one subset and evaluation on another, enabling assessment of the model's generalization performance on unseen data. This procedure aids in detecting overfitting and ensures the model's reliability and effectiveness in real-world scenarios.

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data, y_train, y_test = train_test_split(scaled_data, blc_target, test_size=0.3, random_state=42)
```

```python
print(train_data.shape)
print(y_train.shape)
```

```python
# Function to plot confusion matrix visually
def plot_cmatrix(con_matrix):
    class_name=['Normal','Suspect','Pathological']
    f, ax=plt.subplots(figsize=(5,5))
    sns.heatmap(con_matrix, annot=True, linewidths=1, linecolor="orange",cmap='Reds', fmt= ".0f", ax=ax,
               xticklabels=class_name, yticklabels=class_name)
    plt.xlabel('Predicted values')
    plt.ylabel('Actual values')
    plt.title('Confusion Matrix')
    plt.show()
```

The plot_cmatrix function is designed to visualize confusion matrices, which are widely used in evaluating the performance of classification models. By plotting the confusion matrix, we can visually inspect how well the model predicts each class.
The function enhances interpretability by annotating the heatmap with the actual and predicted class labels. This visualization helps in identifying patterns of correct and incorrect predictions, such as false positives and false negatives.

# Methods:
## Decision Tree Classifier Model:

```python
##Import libraries for implementing classification algorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

##Import library for evaluating model performance
from sklearn.metrics import confusion_matrix

##Import library for measuring classification accuracy
from sklearn.metrics import accuracy_score
```

```python
# Decision Tree classifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(train_data, y_train)
dt_pred = dt_classifier.predict(test_data)

#For confusion matrix
dt_cmatrix = confusion_matrix(y_test, dt_pred)
print('Decision Tree Classifier confusion matrix :')
plot_cmatrix(dt_cmatrix)
```

The above confusion matrix Interprets the True negative(TN), False Positive (FP), False Negative (FN). 
The model correctly predicts the diagonal points for all three categories which is meant by true positive. For false positive which is denoted as FP is the addition of the row matrix without adding the diagonal for incorrected prediction, while False Negative is column addition across the matrix without adding the diagonal matrix which were incorrectly predicted as another class




```python
#Calculating the accuracy of training data 

df_train_pred = dt_classifier.predict(train_data)
df_train_accurancy = accuracy_score(y_train, df_train_pred)
```

```python
dt_accuracy=accuracy_score(y_test, dt_pred)
print(f"test- \n Accuracy of decision tree classifier is:',{dt_accuracy}")
print(f"train- \n Accuracy of decision tree classifier is:',{df_train_accurancy}")
```

## Random Forest Classifier Model:

```python
# Random Forest Classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(train_data, y_train)
rf_pred = rf_classifier.predict(test_data)
rf_cmatrix = confusion_matrix(y_test, rf_pred)
print('Random Forest Classifier confusion matrix :')
plot_cmatrix(rf_cmatrix)
```

```python
#Calculating the accuracy of training data 

rf_train_pred = rf_classifier.predict(train_data)
rf_train_accurancy = accuracy_score(y_train, rf_train_pred)
```

```python
rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"test- \n Accuracy of decision tree classifier is:',{rf_accuracy}")
print(f"train- \n Accuracy of decision tree classifier is:',{rf_train_accurancy}")
```
